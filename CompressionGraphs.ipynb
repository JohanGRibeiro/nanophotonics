Let's get a feel for the compression sizes for a metasurface of diameter 1 cm for library (referencing) algorithm and direct (non-referencing) algorithm,
both in log and linear graphing.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
nanopillar_counts = [10, 10**2, 10**3, 10**4, 10**5, 10**6]
no_refs_file_sizes = [43.5, 266, 1567, 9322, 58445, 387226]  # in KB
refs_file_sizes = [41, 35, 55, 364, 3525, 35163]  # in KB
error = np.full(6, 0.25)  # Error in KB


# Plot
plt.figure(figsize=(8, 6))
plt.errorbar(nanopillar_counts, no_refs_file_sizes, yerr=error, label="No References", marker='o', capsize=5)
plt.errorbar(nanopillar_counts, refs_file_sizes, yerr=error, label="With References", marker='s', capsize=5)


# Logarithmic scale for x-axis and y-axis
plt.xscale("log")


# Labels and legend
plt.xlabel("Number of Nanopillars (log scale)")
plt.ylabel("File Size (KB, log scale)")
plt.title("File Sizes vs Number of Nanopillars (log)")
plt.legend()
plt.grid(which="both", linestyle="--", linewidth=0.5)

# Display plot
```

Now let's see the difference of each method for each number of nanopillars:

```python
# Data
nanopillar_counts = [10, 10**2, 10**3, 10**4, 10**5, 10**6]
no_refs_file_sizes = [43.5, 266, 1567, 9322, 58445, 387226]  # in KB
refs_file_sizes = [41, 35, 55, 364, 3525, 35163]  # in KB

# Compute the differences
differences = [no_refs - refs for no_refs, refs in zip(no_refs_file_sizes, refs_file_sizes)]

# Print the differences for clarity
for np, diff in zip(nanopillar_counts, differences):
    print(f"Nanopillars: {np}, Size Difference (No Refs - With Refs): {diff:.2f} KB")

# Plot the differences
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(nanopillar_counts, differences, marker='o', linestyle='-', color='red', label="Size Difference")

# Logarithmic scale for x-axis
plt.xscale("log")

# Labels and legend
plt.xlabel("Number of Nanopillars (log scale)")
plt.ylabel("Size Difference (KB)")
plt.title("File Size Difference: No References vs With References")
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)  # Reference line at 0
plt.legend()
plt.grid(which="both", linestyle="--", linewidth=0.5)

# Display plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
nanopillars = [10, 10**2, 10**3, 10**4, 10**5, 10**6]
no_refs_file_sizes = [43.5, 266, 1567, 9322, 58445, 387226]  # kB
refs_file_sizes = [41, 35, 55, 364, 3525, 35163]  # kB
error = 0.25  # Error in KB

# Plot
plt.figure(figsize=(10, 6))
plt.errorbar(nanopillar_counts, no_refs_file_sizes, yerr=error, label="No References", marker='o', capsize=5, linestyle='-')
plt.errorbar(nanopillar_counts, refs_file_sizes, yerr=error, label="With References", marker='s', capsize=5, linestyle='--')
plt.plot(nanopillars, no_refs_file_sizes, label="Without References", marker="o", linestyle="-")
plt.plot(nanopillars, refs_file_sizes, label="With References", marker="s", linestyle="--")

# Labels and legend
plt.title("File Sizes vs Number of Nanopillars (linear)", fontsize=14)
plt.xlabel("Number of Nanopillars (linear)", fontsize=12)
plt.ylabel("File Size (kB) (linear)", fontsize=12)
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
